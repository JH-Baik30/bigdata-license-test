2 유튜브 공범컨텐츠 동영상

```python
import pandas as pd

channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

각 데이터의 ‘ct’컬럼을 시간으로 인식할수 있게 datatype을 변경하고\
video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라

```python
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])
```

```python
video['videoname'].value_counts()
```

수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라

```python
# 내 답안은 결괏값의 정렬이 안되어있음, 마지막에 정렬을 붙여봄 어차피 drop 할때 마지막값을 차용하나, 데이터가 뒤죽박죽이라면 문제가 된다.
video[['viewcnt','videoname','ct']].drop_duplicates('videoname', keep='last').reset_index(drop=True).sort_values(by='videoname')
```

```python
# 1 data 정렬, 2 마지막값 남기고 중복 제거, 3 필요한 컬럼만 남기기, 4 인덱스 정리
video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last')[['viewcnt','ct','videoname']].reset_index(drop=True)
```

Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

```python
channel[channel['ct'] >= '2021-10-03'].sort_values(['ct','channelname'])\
    .drop_duplicates('channelname', keep='first')[['channelname','subcnt']].reset_index(drop=True)
```

각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
end_df = channel[channel['ct'].dt.strftime('%Y-%m-%d %H') == '2021-11-01 15']
end_df = end_df[['channelname','subcnt']].reset_index(drop=True)
end_df.columns = ['channelname', 'end_sub']

start_df = channel[channel['ct'].dt.strftime('%Y-%m-%d %H') == '2021-10-03 03']
start_df = start_df[['channelname','subcnt']].reset_index(drop=True)
start_df.columns = ['channelname','start_sub']
```

```python
tt = pd.merge(start_df, end_df)
tt['del'] = tt['end_sub'] - tt['start_sub']
tt[['channelname','del']]
```

각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다. \
공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라

```python
import datetime
ep1 = video[video['videoname'].str.contains('EP1')].sort_values('ct').reset_index(drop=True)
ep1[(ep1.ct.diff(1) >= datetime.timedelta(minutes=20)) | (ep1.ct.diff(1) <= datetime.timedelta(5))]

# 도무지 뭔소린지..
```

각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라

```python
start_date = video.sort_values(['ct','videoname']).drop_duplicates('videoname', keep='first')[['ct','videoname']]
start_date['date'] = start_date['ct'].dt.date
start_date[['date','videoname']]
```

“공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.\ 
공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
video['time'] = video['ct'].dt.hour
answer = video.sort_values(['videoname','ct']).loc[video['time'] == 21].drop_duplicates('videoname')
answer[['viewcnt','ct','videoname']].sort_values('viewcnt', ascending=False).reset_index(drop=True)
```

```python
video.drop(['time'], axis=1, inplace=True)
```

video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 \
videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

```python
video.head(3)
```

```python
target = video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last')
target['ratio'] = video['dislikecnt'] / video['likecnt']
target[['videoname','ratio']].sort_values('ratio')
```

2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
video.head(3)
```

```python
start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')
video[(video['ct'] >= start)&(video['ct'] <= end)][['videoname','viewcnt']].groupby('videoname').agg(max) -\
    video[(video['ct'] >= start)&(video['ct'] <= end)][['videoname','viewcnt']].groupby('videoname').agg(min)
```

```python
start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')
test = video[(video['ct'] >= start)&(video['ct'] <= end)][['videoname','viewcnt']].groupby('videoname')
test.agg(max) - test.agg(min)
```
